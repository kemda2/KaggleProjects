```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.base import clone
from sklearn.feature_selection import mutual_info_classif, f_classif
from scipy.stats import chi2_contingency

from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold
from category_encoders import TargetEncoder

def full_eda(name, df, target, id_col=None, sample_size_plots=10000, 
             random_state=42, run_null_importance=True,
             n_shuffles=2, n_splits=3, save_plots=True, show_plots=False):
    """
    Kapsamlı EDA + Null Importance Analizi

    Parametreler
    -----------
    df : pd.DataFrame
        Analiz edilecek veri seti.
    target : str
        Hedef değişkenin sütun adı (Test setinde yoksa None verilebilir).
    id_col : str, optional
        Eğer veri setinde bir ID sütunu varsa adı; analiz öncesi çıkarılır.
    sample_size_plots : int, default 10000
        Violin plot için örneklem büyüklüğü.
    random_state : int, default 42
        Rastgelelik kontrolü.
    run_null_importance : bool, default True
        Null importance analizi yapılsın mı? (Sadece hedef varsa)
    n_shuffles : int, default 2
        Null importance için shuffle sayısı.
    n_splits : int, default 3
        Null importance CV kat sayısı.
    save_plots : bool, default True
        Grafikler diske kaydedilsin mi?
    show_plots : bool, default False
        Grafikler ekranda gösterilsin mi? (Notebook için True yapabilirsiniz)
    """

    # ID sütununu çıkar
    if id_col is not None and id_col in df.columns:
        df = df.drop(columns=[id_col])

    # Temel bilgiler
    print("="*80)
    print("1. VERİ SETİ GENEL BİLGİLERİ")
    print("="*80)
    print(f"Shape: {df.shape}")
    print(f"Duplicate rows: {df.duplicated().sum()}")
    print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    print(f"\nİlk 5 satır:\n{df.head().to_string()}\n")
    df.info()

    # Eksik değer analizi
    print("\n" + "="*80)
    print("2. EKSİK DEĞER ANALİZİ")
    print("="*80)
    missing = df.isnull().sum()
    missing_pct = (missing / len(df)) * 100
    missing_df = pd.DataFrame({'Count': missing, 'Percentage': missing_pct})
    missing_df = missing_df[missing_df['Count'] > 0].sort_values('Count', ascending=False)
    if len(missing_df) == 0:
        print("✅ Hiç eksik değer yok!")
    else:
        print(missing_df.to_string())

    # Sütun türlerini otomatik belirle
    num_cols = df.select_dtypes(include=np.number).columns.tolist()
    cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

    if target is not None and target in num_cols:
        num_cols.remove(target)
    if target is not None and target in cat_cols:
        cat_cols.remove(target)

    print("\n" + "="*80)
    print("3. SÜTUN BİLGİLERİ")
    print("="*80)
    col_info = pd.DataFrame({
        'Type': ['Numerical']*len(num_cols) + ['Categorical']*len(cat_cols),
        'Unique': [df[c].nunique() for c in num_cols] + [df[c].nunique() for c in cat_cols],
        'Dtype': [df[c].dtype for c in num_cols] + [df[c].dtype for c in cat_cols]
    }, index=num_cols + cat_cols)
    print(col_info.to_string())

    if target is None or target not in df.columns:
        print("\n⚠️ Hedef değişken bulunamadı. Sadece yapısal analiz tamamlandı.")
        return

    # Hedef dağılımı
    print("\n" + "="*80)
    print("4. HEDEF SINIF DAĞILIMI")
    print("="*80)
    target_counts = df[target].value_counts().sort_index()
    target_pcts = ((df[target].value_counts(normalize=True).sort_index())*100).round(2)
    target_stats = pd.DataFrame({
        'Count': target_counts,
        'Percentage': target_pcts,
        'Imbalance_Ratio': (target_counts.max() / target_counts).round(2)
    })
    print(target_stats)

    # Sayısal değişken analizleri
    if len(num_cols) > 0:
        print("\n" + "="*80)
        print("5. SAYISAL DEĞİŞKENLER – SINIF BAZINDA İSTATİSTİKLER & GRAFİKLER")
        print("="*80)
        for col in num_cols:
            print(f"\n📌 {col}")
            # İstatistikler
            group_stats = df.groupby(target)[col].describe(
                percentiles=[0.01, 0.05, 0.10, 0.25, 0.50, 0.75, 0.90, 0.95, 0.99]
            ).round(3)
            print(group_stats.to_string())

            # Grafikler
            fig, axes = plt.subplots(1, 3, figsize=(24, 6))
            sns.histplot(data=df, x=col, hue=target, bins=40, kde=True, ax=axes[0], palette="Set2")
            axes[0].set_title(f'{col} Distribution by Class', fontsize=12, fontweight='bold')
            sns.boxplot(data=df, x=target, y=col, hue=target, ax=axes[1], palette="Set2")
            axes[1].set_title(f'{col} Boxplot by Class', fontsize=12, fontweight='bold')
            # düzeltilmiş violin plot
            sns.violinplot(data=df.sample(min(sample_size_plots, len(df)), random_state=random_state),
                           x=target, y=col, hue=target, legend=False, palette='Set2', ax=axes[2])
            axes[2].set_title(f'{col} Violin Plot', fontsize=12, fontweight='bold')
            plt.tight_layout()
            if save_plots:
                plt.savefig(f"{name}_eda_{col}.png", dpi=300, bbox_inches='tight')
            if show_plots:
                plt.show()
            else:
                plt.close()

    # Kategorik değişken analizleri
    if len(cat_cols) > 0:
        print("\n" + "="*80)
        print("6. KATEGORİK DEĞİŞKENLER – SINIF BAZINDA ANALİZ")
        print("="*80)
        for col in cat_cols:
            print(f"\n📌 {col}")
            cat_stats = df.groupby(col).agg(
                count=(col, 'count'),
                target_mean=(target, 'mean'),
                target_std=(target, 'std')
            ).round(4)
            cat_stats['count_pct'] = (cat_stats['count'] / len(df) * 100).round(2)
            cat_stats = cat_stats.sort_values('target_mean', ascending=False)
            print(cat_stats.to_string())

            # Grafikler
            fig, axes = plt.subplots(1, 3, figsize=(27, 8))
            cat_stats['count'].plot(kind='bar', ax=axes[0], color='steelblue', edgecolor='black')
            axes[0].set_title(f'{col} Frequency', fontsize=12, fontweight='bold')
            axes[0].tick_params(axis='x', rotation=45)

            colors_bar = ['#e74c3c' if x > 0.5 else '#2ecc71' if x < 0.3 else '#f39c12'
                          for x in cat_stats['target_mean']]
            cat_stats['target_mean'].plot(kind='bar', ax=axes[1], color=colors_bar, edgecolor='black')
            axes[1].axhline(y=df[target].mean(), color='black', linestyle='--', alpha=0.5, label='Global Mean')
            axes[1].legend()
            axes[1].tick_params(axis='x', rotation=45)

            crosstab_norm = pd.crosstab(df[col], df[target], normalize='index')
            crosstab_norm.plot(kind='bar', stacked=True, ax=axes[2], colormap='Set2', edgecolor='black')
            axes[2].set_title(f'{col} Class Distribution', fontsize=12, fontweight='bold')
            axes[2].tick_params(axis='x', rotation=45)
            plt.tight_layout()
            if save_plots:
                plt.savefig(f"{name}_eda_{col}.png", dpi=300, bbox_inches='tight')
            if show_plots:
                plt.show()
            else:
                plt.close()

    # Korelasyon matrisi (sadece sayısal + hedef)
    if len(num_cols) > 0:
        print("\n" + "="*80)
        print("7. KORELASYON ANALİZİ")
        print("="*80)
        corr_matrix = df[num_cols + [target]].corr(numeric_only=True)
        plt.figure(figsize=(12, 10))
        mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)
        sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r',
                    center=0, square=True, linewidths=0.5, cbar_kws={'shrink': 0.8})
        plt.title('Correlation Matrix (Numerical + Target)', fontsize=14, fontweight='bold')
        plt.tight_layout()
        if save_plots:
            plt.savefig(f"{name}_correlation_matrix.png", dpi=300, bbox_inches='tight')
        if show_plots:
            plt.show()
        else:
            plt.close()

        high_corr = []
        for i in range(len(corr_matrix.columns)):
            for j in range(i+1, len(corr_matrix.columns)):
                if abs(corr_matrix.iloc[i, j]) > 0.7:
                    high_corr.append({
                        'Feature1': corr_matrix.columns[i],
                        'Feature2': corr_matrix.columns[j],
                        'Correlation': corr_matrix.iloc[i, j]
                    })
        if high_corr:
            print("\n⚠️ YÜKSEK KORELASYONLU ÇİFTLER (|r| > 0.7):")
            print(pd.DataFrame(high_corr).sort_values('Correlation', ascending=False).to_string())
        else:
            print("\n✅ Yüksek korelasyonlu çift yok (|r| > 0.7)")

    # Tek değişkenli önem analizi
    print("\n" + "="*80)
    print("8. TEK DEĞİŞKENLİ ÖNEM ANALİZİ")
    print("="*80)
    if len(num_cols) > 0:
        num_importance = pd.DataFrame(index=num_cols)
        f_vals, p_vals = [], []
        for col in num_cols:
            f, p = f_classif(df[[col]].fillna(df[col].median()), df[target])
            f_vals.append(f[0])
            p_vals.append(p[0])
        num_importance["F_value"] = f_vals
        num_importance["F_pvalue"] = p_vals
        num_importance["Mutual_Info"] = mutual_info_classif(
            df[num_cols].fillna(df[num_cols].median()), df[target], random_state=random_state
        )
        num_importance = num_importance.sort_values("Mutual_Info", ascending=False)
        print("\n📈 Sayısal Değişken Önem Sıralaması:")
        print(num_importance.round(4).to_string())

    if len(cat_cols) > 0:
        cat_importance = pd.DataFrame(index=cat_cols)
        chi2_stats = []
        for col in cat_cols:
            contingency = pd.crosstab(df[col], df[target])
            chi2, p, dof, _ = chi2_contingency(contingency)
            chi2_stats.append([chi2, p])
        cat_importance["Chi2"] = [x[0] for x in chi2_stats]
        cat_importance["p_value"] = [x[1] for x in chi2_stats]
        cat_importance = cat_importance.sort_values("Chi2", ascending=False)
        print("\n📊 Kategorik Değişken Önem Sıralaması:")
        print(cat_importance.round(4).to_string())

    # Görselleştirme (önem)
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    if len(num_cols) > 0:
        axes[0].barh(num_importance.index[::-1], num_importance["Mutual_Info"][::-1], color="steelblue")
        axes[0].set_title("Numerical Features - Mutual Information")
    if len(cat_cols) > 0:
        axes[1].barh(cat_importance.index[::-1], cat_importance["Chi2"][::-1], color="coral")
        axes[1].set_title("Categorical Features - Chi-Square")
    plt.tight_layout()
    if save_plots:
        plt.savefig(f"{name}_univariate_importance.png", dpi=300, bbox_inches='tight')
    if show_plots:
        plt.show()
    else:
        plt.close()

    # Null Importance Analizi (opsiyonel)
    if run_null_importance and (len(num_cols) + len(cat_cols)) > 0:
        print("\n" + "="*80)
        print("9. NULL IMPORTANCE ANALİZİ")
        print("="*80)
        X = df[num_cols + cat_cols].copy()
        y = df[target].values
        print(f"Toplam özellik sayısı: {X.shape[1]}")

        def get_model():
            return XGBClassifier(
                n_estimators=100, max_depth=4, learning_rate=0.05,
                tree_method='hist', eval_metric='logloss', random_state=random_state, verbosity=0
            )

        model = get_model()

        # Gerçek önem
        encoder = TargetEncoder(cols=cat_cols, smoothing=0.1)
        X_enc = encoder.fit_transform(X, y)
        model.fit(X_enc, y)
        importance_dict = model.get_booster().get_score(importance_type='gain')
        actual_imp = np.array([importance_dict.get(f'f{i}', 0) for i in range(X.shape[1])])

        # Null önem
        cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)
        null_imps = []
        for shuffle_num in range(n_shuffles):
            y_shuffled = y.copy()
            np.random.shuffle(y_shuffled)
            for tr_idx, _ in cv.split(X, y_shuffled):
                X_tr = X.iloc[tr_idx]
                y_tr = y_shuffled[tr_idx]
                enc = TargetEncoder(cols=cat_cols, smoothing=0.1)
                X_tr_enc = enc.fit_transform(X_tr, y_tr)
                m = clone(model)
                m.fit(X_tr_enc, y_tr)
                imp_dict = m.get_booster().get_score(importance_type='gain')
                null_imps.append([imp_dict.get(f'f{i}', 0) for i in range(X.shape[1])])

        null_imps = np.array(null_imps)
        null_mean = null_imps.mean(axis=0)
        null_95 = np.percentile(null_imps, 95, axis=0)
        safe_score = actual_imp / (null_95 + 1e-10)

        null_importance_df = pd.DataFrame({
            'feature': X.columns,
            'actual_importance': actual_imp,
            'null_mean': null_mean,
            'null_95th': null_95,
            'safe_score': safe_score
        }).sort_values('safe_score', ascending=False)

        print("\n📊 Null Importance Sonuçları (ilk 20):")
        print(null_importance_df[['feature', 'actual_importance', 'null_95th', 'safe_score']].head(20).round(4).to_string())

        significant_features = null_importance_df[null_importance_df['safe_score'] > 1]['feature'].tolist()
        print(f"\n💡 Gerçek sinyal taşıyan özellik sayısı: {len(significant_features)}")
        print("Özellikler:", significant_features if significant_features else "Yok")

        # Grafik
        fig, ax = plt.subplots(figsize=(10, 6))
        top = null_importance_df.head(20)
        ax.barh(top['feature'][::-1], top['safe_score'][::-1],
                color=['#2ecc71' if s > 1 else '#e74c3c' for s in top['safe_score'][::-1]])
        ax.axvline(1, color='black', linestyle='--')
        ax.set_title('Null Importance – Safe Score (>1 = signal)')
        plt.tight_layout()
        if save_plots:
            plt.savefig(f"{name}_null_importance_safe.png", dpi=300, bbox_inches='tight')
        if show_plots:
            plt.show()
        else:
            plt.close()

    print("\n✅ EDA TAMAMLANDI.")






target = "PitNextLap"
idc = "id"

paths = [
    "/kaggle/input/competitions/playground-series-s6e5/train.csv",
    "/kaggle/input/competitions/playground-series-s6e5/test.csv",
    "/kaggle/input/datasets/aadigupta1601/f1-strategy-dataset-pit-stop-prediction/f1_strategy_dataset_v4.csv",
]

train = pd.read_csv(paths[0])
test = pd.read_csv(paths[1])
orig = pd.read_csv(paths[2])

# # Yıl ve PitStop'u stringe çevir (kategorik olsun diye)
# train["Year"], test["Year"] = train["Year"].astype(str), test["Year"].astype(str)
# train["PitStop"], test["PitStop"] = train["PitStop"].astype(str), test["PitStop"].astype(str)

test_id = test[idc]

train = train.drop(columns=[idc], errors="ignore")
test = test.drop(columns=[idc], errors="ignore")
orig = orig.drop(columns=[idc], errors="ignore")
orig = orig[[c for c in train.columns]+["Normalized_TyreLife"]]

full_eda(name="Train", df=train, target=target, id_col=None, sample_size_plots=10000, 
             random_state=42, run_null_importance=True,
             n_shuffles=2, n_splits=3, save_plots=True, show_plots=True)
full_eda(name="Test", df=test, target=None, id_col=None, sample_size_plots=10000, 
             random_state=42, run_null_importance=True,
             n_shuffles=2, n_splits=3, save_plots=True, show_plots=True)
full_eda(name="Original", df=orig, target=target, id_col=None, sample_size_plots=10000, 
             random_state=42, run_null_importance=True,
             n_shuffles=2, n_splits=3, save_plots=True, show_plots=True)


!zip -r eda_results_fixed.zip /kaggle/working -i "*.png"
```
